```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def call_payoff(spot: float, strike: float) -> float:
    return np.maximum(spot - strike, 0)

def put_payoff(spot: float, strike: float) -> float:
    return np.maximum(strike - spot, 0)
```

```python
spots = np.linspace(50, 150, 200)
strike = 100

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].plot(spots, call_payoff(spots, strike))
axes[0].set_title("Long Call Payoff at Expiry (K=100)")

axes[0].axhline(0, color='black', linewidth=0.5)
axes[0].axvline(strike, color='red', linestyle='--', label='Strike')
axes[0].set_xlabel("Spot Price")
axes[0].set_ylabel("Payoff ($)")
axes[0].legend()

axes[1].plot(spots, put_payoff(spots, strike), color='orange')
axes[1].set_title("Put Payoff at Expiry (K=100)")
axes[1].axhline(0, color='black', linewidth=0.5)
axes[1].axvline(strike, color='red', linestyle='--', label='Strike')
axes[1].set_xlabel("Spot Price")
axes[1].set_ylabel("Payoff ($)")
axes[1].legend()

plt.tight_layout()
plt.show()
```
